### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()
```

# 2. Vendas por Categoria

```python
vendasporcatl1 = loja.groupby("Categoria do Produto")['Categoria do Produto'].count()
vendasporcatl2 = loja2.groupby("Categoria do Produto")['Categoria do Produto'].count()
vendasporcatl3 = loja3.groupby("Categoria do Produto")['Categoria do Produto'].count()
vendasporcatl4 = loja4.groupby("Categoria do Produto")['Categoria do Produto'].count()

```

# 3. Média de Avaliação das Lojas


```python
mediaavaliacao_loja1 = loja["Avaliação da compra"].mean()
mediaavaliacao_loja2 = loja2["Avaliação da compra"].mean()
mediaavaliacao_loja3 = loja3["Avaliação da compra"].mean()
mediaavaliacao_loja4 = loja4["Avaliação da compra"].mean()
```

# 4. Produtos Mais e Menos Vendidos

```python
loja.groupby("Produto")['Produto'].count()
```

```python

produtomaisvendidol1 = loja.groupby("Produto")['Produto'].count().idxmax()
produtomaisvendidol2 = loja2.groupby("Produto")['Produto'].count().idxmax()
produtomaisvendidol3 = loja3.groupby("Produto")['Produto'].count().idxmax()
produtomaisvendidol4 = loja4.groupby("Produto")['Produto'].count().idxmax()
```

```python
produtomenosvendidol1 = loja.groupby("Produto")['Produto'].count().idxmin()
produtomenosvendidol2 = loja2.groupby("Produto")['Produto'].count().idxmin()
produtomenosvendidol3 = loja3.groupby("Produto")['Produto'].count().idxmin()
produtomenosvendidol4 = loja4.groupby("Produto")['Produto'].count().idxmin()
```

# 5. Frete Médio por Loja

```python
fretemedl1 = loja["Frete"].mean()
fretemedl2 = loja2["Frete"].mean()
fretemedl3 = loja3["Frete"].mean()
fretemedl4 = loja4["Frete"].mean()
```

6.Visualização

```python
import matplotlib.pyplot as plt

x = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
y = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
plt.bar(x, y)
```

```python
xline = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
yline = [mediaavaliacao_loja1, mediaavaliacao_loja2, mediaavaliacao_loja3, mediaavaliacao_loja4]
plt.plot(xline, yline)
```

```python

x = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
y = [fretemedl1, fretemedl2, fretemedl3, fretemedl4]

plt.figure(figsize=(5, 5))  

plt.pie(
    y,
    labels=x,
    autopct="%1.1f%%",  
    startangle=90    
)

plt.title("Frete Médio por Loja")
plt.tight_layout()
plt.show()
```

## Análise das Lojas

Com base nos gráficos observados, é possível identificar diferentes cenários de desempenho entre as quatro lojas avaliadas. A seguir, apresenta-se uma análise organizada:

---

### **Loja 4 — A mais crítica**
A Loja 4 se destaca negativamente por exigir maior esforço e práticas de empreendedorismo para alcançar:
- Um **faturamento aceitável**  
- **Boas avaliações dos clientes**

➡️ **Conclusão:** Vender a Loja 4 é a melhor opção estratégica, considerando o baixo desempenho geral e o alto esforço necessário para torná-la competitiva.

---

### **Loja 1 — Boa em vendas, ruim em avaliações**
A Loja 1:
- Lidera em **volume de vendas**
- Também lidera em **avaliações negativas**

É necessário investigar as causas dessas avaliações negativas. Observando o **gráfico de frete**, é possível que grande parte dessas reclamações esteja relacionada ao custo do frete.

➡️ **Ação recomendada:**
- Confirmar se o frete é realmente o principal motivo das avaliações negativas  
- Caso seja confirmado, **adotar medidas para reduzir o frete**  
- Isso deve melhorar a experiência do cliente, evitar perda de consumidores e proteger o lucro no longo prazo

---

### **Lojas 2 e 3 — Equilíbrio saudável**
As Lojas 2 e 3 apresentam:
- **Bom faturamento**
- **Boas avaliações**
- **Comportamento equilibrado** nos indicadores

➡️ **Conclusão:** Não são necessárias grandes ações corretivas neste momento. Elas operam de maneira estável e com bom desempenho.

---

## ✔️ Resumo Geral
- **Vender Loja 4** → Melhor decisão estratégica devido ao baixo desempenho e alto custo de recuperação  
- **Ajustar a Loja 1** → Focar na redução do frete e investigação das reclamações  
- **Manter Lojas 2 e 3** → Continuar monitorando, mas não exigem mudanças significativas  

## Atividade Extra


```python

vendas_por_ponto = (
    loja.groupby(['lat', 'lon'])
      .size()
      .reset_index(name='vendas')
)


plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    vendas_por_ponto['lon'],   # eixo X: longitude
    vendas_por_ponto['lat'],   # eixo Y: latitude
    s=vendas_por_ponto['vendas'] * 10,  
    c=vendas_por_ponto['vendas'],
    cmap='viridis',
    alpha=0.7,
    edgecolor='k',
    linewidth=0.5
)

plt.colorbar(scatter, label='Vendas (contagem por coordenada)')
plt.title('Vendas por Localidade (bolhas proporcionais à contagem)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
